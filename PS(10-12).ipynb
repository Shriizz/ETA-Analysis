```python
census_data = pd.read_csv("C:/Users/Shrii/Documents/Internship Program/Internship Codes/Census/Clean_data/census.csv")
```

```python
census_data.dropna(subset=['Population'], inplace=True)
```

```python
census_data['Population'] = census_data['Population'].astype(float)
```

```python
census_data['Population']
```

```python
state_population = census_data.groupby('State/UT')['Population'].sum()
```

```python
merged_data = pd.merge(hospital_data[['State/UT', 'Total_Hospital_Beds']], state_population,on='State/UT')
```

```python
merged_data.isnull().sum()
```

```python
merged_data['Total_Hospital_Beds'] = merged_data['Total_Hospital_Beds'].astype(float)
```

```python
merged_data['Beds_Per_10,000'] = (merged_data['Total_Hospital_Beds'] / merged_data['Population']) * 10000
```

```python
merged_data.head()
```

```python
national_average = merged_data['Total_Hospital_Beds'].sum() / merged_data['Population'].sum() * 10000
```

```python
plt.figure(figsize=(10, 6))
plt.barh(merged_data['State/UT'], merged_data['Beds_Per_10,000'], color='purple')
plt.axvline(x=national_average, color='blue', linestyle='--', label='National Average')
plt.xlabel('Beds_Per_10,000 People')
plt.ylabel('State/UT')
plt.title('Hospital Beds Per 10,000 People by State/UT')
plt.legend()
plt.tight_layout()

plt.show()
```
